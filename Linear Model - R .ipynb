```r
plot(x = csv_data$x,
     y = csv_data$y,
     xlab = "x",
     ylab = "y",
     col = "black",
     main = "Linear Model2 - R")
  
abline(lm(csv_data$x ~ csv_data$y,
          data = csv_data), col = "black")
```
